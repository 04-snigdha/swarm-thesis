# Swarm Thesis — Simulation Analysis
Decentralized Collective Transport via Mechanical & Chemical Stigmergy

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_theme(style="whitegrid")

df = pd.read_csv("results.csv")
print(f"Loaded {len(df)} trials")
df.head()
```

```python
# Quick overview
print("Shapes:", df['Shape'].unique())
print("Swarm sizes:", sorted(df['Swarm_Size'].unique()))
print("\nOverall success rate:", f"{df['Success'].mean()*100:.1f}%")
print("\nTrials per combination:")
df.groupby(['Shape', 'Swarm_Size'])['Success'].count().unstack()
```

## 1. Success Rate vs Swarm Size

```python
success_rates = df.groupby(['Shape', 'Swarm_Size'])['Success'].mean().mul(100).reset_index()
success_rates.columns = ['Shape', 'Swarm_Size', 'Success_Rate']

fig, ax = plt.subplots(figsize=(10, 6))
sns.lineplot(data=success_rates, x='Swarm_Size', y='Success_Rate', hue='Shape', marker='o', ax=ax)
ax.set_title('Success Rate vs Swarm Size', fontsize=14)
ax.set_xlabel('Swarm Size (N)')
ax.set_ylabel('Success Rate (%)')
ax.set_ylim(0, 105)
plt.tight_layout()
plt.savefig('success_rate_vs_swarm_size.png', dpi=150)
plt.show()
```

## 2. Completion Time Distribution (Successful Trials)

```python
successful = df[df['Success'] == True]

fig, ax = plt.subplots(figsize=(10, 6))
sns.boxplot(data=successful, x='Shape', y='Completion_Time',
            order=['Circle', 'Square', 'L-shape', 'C-shape'], ax=ax)
ax.set_title('Completion Time Distribution by Shape (Successful Trials)', fontsize=14)
ax.set_xlabel('Payload Shape')
ax.set_ylabel('Completion Time (s)')
plt.tight_layout()
plt.savefig('completion_time_distribution.png', dpi=150)
plt.show()

successful.groupby('Shape')['Completion_Time'].describe().round(2)
```

## 3. Completion Time by Swarm Size and Shape

```python
fig, ax = plt.subplots(figsize=(12, 6))
sns.boxplot(data=successful, x='Swarm_Size', y='Completion_Time', hue='Shape',
            order=sorted(successful['Swarm_Size'].unique()), ax=ax)
ax.set_title('Completion Time by Swarm Size and Shape', fontsize=14)
ax.set_xlabel('Swarm Size (N)')
ax.set_ylabel('Completion Time (s)')
plt.tight_layout()
plt.show()
```

## 4. Jamming / Failure Frequency

```python
failures = df[df['Success'] == False]
jamming_freq = failures.groupby(['Shape', 'Swarm_Size']).size().reset_index(name='Failures')

fig, ax = plt.subplots(figsize=(10, 6))
sns.barplot(data=jamming_freq, x='Shape', y='Failures', hue='Swarm_Size',
            order=['Circle', 'Square', 'L-shape', 'C-shape'], ax=ax)
ax.set_title('Failed Trials by Shape and Swarm Size', fontsize=14)
ax.set_xlabel('Payload Shape')
ax.set_ylabel('Number of Failed Trials')
plt.tight_layout()
plt.savefig('jamming_frequency.png', dpi=150)
plt.show()
```

## 5. Success Rate Heatmap

```python
heatmap_data = df.groupby(['Shape', 'Swarm_Size'])['Success'].mean().mul(100).unstack()
heatmap_data = heatmap_data.reindex(['Circle', 'Square', 'L-shape', 'C-shape'])

fig, ax = plt.subplots(figsize=(8, 5))
sns.heatmap(heatmap_data, annot=True, fmt='.1f', cmap='RdYlGn',
            vmin=0, vmax=100, ax=ax, cbar_kws={'label': 'Success Rate (%)'})
ax.set_title('Success Rate (%) — Shape × Swarm Size', fontsize=14)
ax.set_xlabel('Swarm Size (N)')
ax.set_ylabel('Payload Shape')
plt.tight_layout()
plt.show()
```

## 6. Thesis Summary

```python
mean_times = successful.groupby(['Shape', 'Swarm_Size'])['Completion_Time'].mean().round(2).unstack()
mean_times = mean_times.reindex(['Circle', 'Square', 'L-shape', 'C-shape'])

print("=== Mean Completion Time (s) — Successful Trials ===")
print(mean_times.to_string())
print()
print("=== Success Rate (%) ===")
print(heatmap_data.to_string())
```
